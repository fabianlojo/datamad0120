```python
import re
import pandas as pd
import numpy as np

df = pd.read_csv("GSAF5.csv", encoding="ISO-8859-1")
df
#(encoding="ISO-8859-1") = encodes what it refers to as "Latin alphabet no. 1", consisting of 
#191 characters from the Latin script. This character-encoding scheme is used throughout the Americas,
#Western Europe, Oceania, and much of Africa. It is also commonly used in most standard romanizations 
#of East-Asian languages. With this encode we are able to read this and many other files.
```

```python
#thinking about our hypothesis where we will try to demonstrate that shark's attacks on human 
#beings are increasing because humans invade their habitat more and more for different reasons.

#first we will start by cleaning our data and get from it the most important we think we will need
#for our investigation.

#Creating new DataFrame without duplicated columns or innecesary for our investigation:
dfa = df.drop(columns=["Case Number", "pdf", "href formula", "href", "Case Number.1", "Case Number.2",
                       "Unnamed: 22", "Unnamed: 23"])

#rearranging DataFrame:
#change columns order:
cols=list(dfa.columns.values)
dfa = dfa[[cols[-1]] + cols[0:-1]]

#change column Name:
x = {"original order":"Case Number"}
dfa = dfa.rename(columns=x)

#define specific data to work on our hypothesis:
dfa.loc[dfa["Type"]=="Boat", "Type"] = "Unprovoked"
dfa.loc[dfa["Type"]=="Boating", "Type"] = "Unprovoked"
dfa.loc[dfa["Type"]=="Sea Disaster", "Type"] = "Provoked"
dfa.loc[dfa["Type"]=="Invalid", "Type"] = "Unknown"
dfa.loc[dfa["Type"]==" ", "Type"] = "Unknown"
dfa.loc[dfa["Type"]=="NaN", "Type"] = "Unknown"
dfa.loc[dfa["Name"]=="male", "Name"] = "Unknown"
dfa.loc[dfa["Name"]=="female", "Name"] = "Unknown"
dfa.loc[dfa["Name"]=="boy", "Name"] = "Unknown"
dfa.loc[dfa["Name"]==" ", "Name"] = "Unknown"
dfa.loc[dfa["Name"]=="NaN", "Name"] = "Unknown"
dfa.loc[dfa["Activity"]==" ", "Activity"] = "Unknown"
dfa.loc[dfa["Activity"]=="NaN", "Activity"] = "Unknown"
dfa.loc[dfa["Case Number"]==" ", "Case Number"] = "Unknown"
dfa.loc[dfa["Case Number"]=="NaN", "Case Number"] = "Unknown"
dfa.loc[dfa["Date"]==" ", "Date"] = "Unknown"
dfa.loc[dfa["Date"]=="NaN", "Date"] = "Unknown"
dfa.loc[dfa["Year"]==" ", "Year"] = "Unknown"
dfa.loc[dfa["Year"]=="NaN", "Year"] = "Unknown"
dfa.loc[dfa["Country"]==" ", "Country"] = "Unknown"
dfa.loc[dfa["Country"]=="NaN", "Country"] = "Unknown"
dfa.loc[dfa["Area"]==" ", "Area"] = "Unknown"
dfa.loc[dfa["Area"]=="NaN", "Area"] = "Unknown"
dfa.loc[dfa["Location"]==" ", "Location"] = "Unknown"
dfa.loc[dfa["Location"]=="NaN", "Location"] = "Unknown"
#dfa.loc[dfa["Sex"]==" ", "Sex"] = "Unknown"
dfa.loc[dfa["Age"]==" ", "Age"] = "Unknown"
dfa.loc[dfa["Age"]=="NaN", "Age"] = "Unknown"
dfa.loc[dfa["Injury"]==" ", "Injury"] = "Unknown"
dfa.loc[dfa["Injury"]=="NaN", "Injury"] = "Unknown"
dfa.loc[dfa["Fatal (Y/N)"]==" ", "Fatal (Y/N)"] = "Unknown"
dfa.loc[dfa["Fatal (Y/N)"]=="NaN", "Fatal (Y/N)"] = "Unknown"
dfa.loc[dfa["Time"]=="NaN", "Time"] = "Unknown"
dfa.loc[dfa["Time"]==" ", "Time"] = "Unknown"
#dfa.loc[dfa["Species"]=="NaN", "Species"] = "Unknown"
dfa.loc[dfa["Investigator or Source"]==" ", "Investigator or Source"] = "Unknown"
dfa.loc[dfa["Investigator or Source"]=="NaN", "Investigator or Source"] = "Unknown"



#

dfa.groupby(["Type"]).mean().sort_values("Year", ascending=False)
```
